```python
#loss = 0 gelijk draw 1 win 3

#Aantal wedstrijden in dat seizoen.
#Aantal punten behaald in dat seizoen.
#Gemiddeld aantal punten per wedstrijd.
#Winst-, verlies- en gelijk percentage in uitwedstrijden en thuiswedstrijden.
def team_season(team,season):
    teamapid = team_df.loc[(team_df['team_long_name'] == team)]['team_api_id'].values[0]

    result1 = match_df.loc[(match_df['away_team_api_id'] == teamapid)& (match_df['season'] == season)  | (match_df['home_team_api_id'] == teamapid) & (match_df['season'] == season)]
    aantal_wed = result1.shape[0]
    
    home = match_df.loc[(match_df['home_team_api_id'] == teamapid)& (match_df['season'] == season)]
    away= match_df.loc[(match_df['away_team_api_id'] == teamapid)& (match_df['season'] == season)]
    home['punten'] = np.where(home['home_team_goal'] > home['away_team_goal'], 3, np.where(home['home_team_goal'] == home['away_team_goal'], 1, 0))
    away['punten'] = np.where(away['home_team_goal'] < away['away_team_goal'], 3, np.where(away['home_team_goal'] == away['away_team_goal'], 1, 0))
    
    awaysum = away['punten'].sum()
    homesum = home['punten'].sum()
    total_punten = awaysum + homesum
    
    home['outcome']= np.where(home['punten'] ==3, 'W', np.where(home['punten'] == 1, 'D', 'L'))
    away['outcome']= np.where(away['punten'] ==3, 'W', np.where(away['punten'] == 1, 'D', 'L'))
    awayperc = away['outcome'].value_counts()
    homeperc = home['outcome'].value_counts()
    homewinperc = round((homeperc['W'] / 17) * 100, 2)
    homedrawperc = round((homeperc['D'] / 17) * 100, 2)
    homelossperc = round((homeperc['L'] / 17) * 100, 2)
    awaywinperc = round((awayperc['W'] / 17) * 100, 2)
    awaydrawperc = round((awayperc['D'] / 17) * 100, 2)
    awaylossperc = round((awayperc['L'] / 17) * 100, 2)
    
    gem_punten = total_punten/aantal_wed
    return print('Aantal wedstrijden: ' + str(aantal_wed)+ " \nAantal punten: " + str(total_punten) + ' \nGemiddelde punten per wedstrijd ' + str(gem_punten) +'\n \nWin uit percentage ' + str(awaywinperc) +'%' +'\nGelijk uit percentage ' + str(awaydrawperc) +'%'+'\nVerlies uit percentage ' + str(awaylossperc) +'%'+'\n \nWin thuis percentage ' + str(homewinperc) +'%' +'\nGelijk thuis percentage ' + str(homedrawperc) +'%'+'\nVerlies thuis percentage ' + str(homelossperc) +'%')

team_season('Feyenoord','2008/2009')
```
